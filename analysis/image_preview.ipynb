```python
from matplotlib import pyplot as plt
import numpy as np
import PIL
```

```python
data_infile = open('images/image__d0.5m_a15um_p62um_step-4.2_dith23.25_th1_10000eV_2.raw','rb')
data_array = np.fromfile(data_infile,dtype='float32',count=512*402)
data_image = PIL.Image.frombuffer("F",[512,402],data_array,'raw','F',0,1)
```

```python
img1 = np.array(data_image.getdata())
img1=img1.reshape(402,512) 

fig2=plt.figure(figsize=(12,6))
ax1=fig2.add_subplot(111)
ax1.set_xlabel('X [pixels]', fontsize=13,labelpad=15)
ax1.set_ylabel('Y [pixels]', fontsize=13,labelpad=15)

p1= ax1.imshow(img1, cmap='bone')
cbar = fig2.colorbar(p1)
cbar.set_label('Intensity [counts]')
```

```python
x_profile1 = np.mean(img1,0)

plt.plot(x_profile1)
plt.title(r'Intensity profile for $3 \times 10^{8}$ events')
plt.xlabel('X [Pixels]')
plt.ylabel('Mean Intensity [counts]')
```
